```python
#  data standardization for column "engine_type"
df1["engine_type"] = df1["engine_type"].str.replace('Unknown','unknown')
df1["engine_type"] = df1["engine_type"].str.replace('UNK','unknown')
```

```python
# Confirming data standardization
df1["engine_type"].unique()
```

```python
# Filling the missing values
cols_to_fill = [
    'aircraft_category',
    'purpose_of_flight',
    'injury_severity',
    'engine_type',
    'broad_phase_of_flight',
    'publication_date'
]

df1[cols_to_fill] = df1[cols_to_fill].fillna('unknown')
```

##### I filled the missing values in the above columns with "unknown" for transparency over assumptions

```python
# Confirming the changes have occured
df1.isna().sum()
```

```python
# Filling the remaining missing values
# Creating a function where I can check all complete non null pairs and fill the missing values with the matching pairs
def fill_missing_pairs(df1, col1, col2):
    # Building valid non-null, non-duplicate pairs
    valid_pairs = df1[[col1, col2]].dropna().drop_duplicates()

    # Creating lookup dictionaries
    col1_to_col2 = dict(valid_pairs.values)
    col2_to_col1 = dict(valid_pairs[[col2, col1]].values)

    # Filling the missing values
    df1[col2] = df1[col2].fillna(df1[col1].map(col1_to_col2))
    df1[col1] = df1[col1].fillna(df1[col2].map(col2_to_col1))

    return df1
```

```python
pairs = [
    ('make', 'engine_type'),
    ('make', 'amateur_built'),
    ('make', 'model'),
    ('make', 'number_of_engines'),
    ('model', 'number_of_engines'),
    ('model', 'amateur_built'),
    ('injury_severity', 'total_fatal_injuries'),
    ('injury_severity', 'total_serious_injuries'),
    ('injury_severity', 'total_minor_injuries'),
    ('injury_severity', 'total_uninjured'),
    ('injury_severity', 'aircraft_damage'),
    ('location', 'latitude'),
    ('location', 'longitude'),
    ('country', 'longitude'),
    ('country', 'airport_code'),
    ('country', 'airport_name'),
    ('airport_name', 'latitude'),
    ('airport_name', 'longitude'),
    ('airport_name', 'airport_code'),
    ('airport_code', 'longitude'),
    ('airport_code', 'latitude'),
    ('aircraft_category', 'registration_number'),
    ('number_of_engines', 'engine_type'),
    ('weather_condition', 'report_status'),
    ('weather_condition', 'event_date'),
    ('weather_condition', 'location')
]

for col1, col2 in pairs:
    df1 = fill_missing_pairs(df1, col1, col2)
```

```python
# Loading the clean dataset and creating a dataframe
data = pd.read_csv("clean_dataset.csv")
data.head()
```

## Univariate Analysis


```python
# Using a for loop to create histplots for the numeric columns
num_columns = ['total_fatal_injuries', 'total_serious_injuries',
           'total_minor_injuries', 'total_uninjured', 'number_of_engines']

for col in num_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[col], bins=30, kde=True, color='steelblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

Observations : 
* The distribution of total_fatal_injuries is sharply right-skewed since majority of the incidents are evident in the dominant bar exceeding 50,000 observations at 0. There are smaller but noticeable peaks at 1 and 2 fatalities, suggesting that while fatal injuries are uncommon, single- and dual-fatality incidents do occur with some regularity.

* The distribution of total_serious_injuries right-skewed, with a very high concentration of events resulting in zero serious injuries reflected in a dominant bar with nearly 100,000 occurrences. However smaller peaks appear at 1 through 5 injuries, but frequencies decrease rapidly beyond that. The right tail of the distribution stretches beyond 20, showing that high-serious-injury events are rare.

* The distribution of total_minor_injuries is right-skewed, with the highest frequency of incidents occurring at zero minor injuries. The histogram displays a strong peak at 0 and a sharp decline as the injury count increases. The distribution suggests that most events result in little to no minor injuries, with higher counts being uncommon. 

* The distribution of total_uninjured is heavily right-skewed, with a peak at zero uninjured individuals indicating that in many incidents, no one walked away without an injury. There is a rapid decline in frequency as the number of uninjured rises, and values beyond 100 become increasingly rare suggesting that large-scale non-injury events are uncommon.

* The Distribution of number of engines shows a strongly right-skewed distribution with a peak at 1 indicating that most aircrafts involved in accidents have one engine. This shows multi-engine aircraft accidents are relatively uncommon.


```python
# Using a for loop to create boxplots for the numeric columns
for col in num_columns:
    plt.figure(figsize=(8, 1.5))
    sns.boxplot(x=data[col], color='tomato')
    plt.title(f'Boxplot of {col}')
    plt.show()
```

Observations:
* The boxplot for total_fatal_injuries illustrates a compact distribution with a concentration of values at zero, emphasizing that most incidents resulted in no fatalities.

* The boxplot for total_serious_injuries shows a distribution that is moderately concentrated in the 0 to 10 injury range, with a median very close to zero, reflecting a high frequency of incidents with no serious injuries. However, a number of individual outliers, visible as dots between roughly 10 and 20 injuries, suggest the presence of a small set of higher-severity cases.

* The boxplot for total_minor_injuries reveals a right-skewed distribution with the majority of values concentrated below 25 injuries indicating relatively low variability in most incidents. However the plot also highlights numerous values extending  beyond 75 with the most extreme values approaching 175 minor injuries pointing to rare but notable cases of high-severity incidents.

* The boxplot for total_uninjured displays a right-skewed distribution with a wide spread of values. The central portion of the data is concentrated at the lower end, suggesting that most incidents resulted in a small number of uninjured individuals. However, there are several values extending beyond 300, representing rare but significant non-injury events which are likely associated with larger capacity aircraft where most or all passengers were unharmed.

* The boxplot for number_of_engines reflects a right-skewed distribution, with most data points concentrated on 1 engine confirming that single-engine aircraft  are the most common to accidents. The range and distribution highlight that multi-engine setups are rare.










```python
# Using a for loop to create violin plots for the numeric columns
for col in num_columns:
    plt.figure(figsize=(8, 4))
    sns.violinplot(x=data[col])
    plt.title(f'Violin Plot of {col}', fontsize = 14)
    plt.show()

```

Observations:
* The violin plot of total_fatal_injuries shows a highly concentrated distribution near zero indicating that the vast majority of incidents involved no fatalities. The shape is narrow but slightly flared at 1 through 4, suggesting a small number of cases with low but nonzero fatality counts. 

* The violin plot of total_serious_injuries reveals a heavily right-skewed distribution with the widest concentration being at zero serious injuries. This suggests that majority of incidents involved no serious harm. 

* The violin plot for total_minor_injuries shows a strongly right-skewed distribution, with concentrateion being around zero minor injuries. This shows that most incidents involved little to no minor harm.

* The violin plot for total_uninjured shows a sharply right-skewed distribution, with the widest concentration clustered at zero uninjured individuals. This shows that in many incidents, there were no uninjured individuals recorded. 

* The violin plot for number_of_engines displays a right-skewed distribution, with the density centered at 1 engine. This shows that single-engine aircraft are the most prone to accidents.
